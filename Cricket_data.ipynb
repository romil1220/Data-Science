```python
from bs4 import BeautifulSoup
import requests
from csv import writer

```

```python
url = "https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?class=2;id=6;type=team"
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify
```

```python
lists = soup.find_all('tr', class_="data1")
```

```python
lists
```

```python
with open('cricketers.csv', 'w', encoding='utf8', newline='')as f:
 thewriter = writer(f)
 header = ['Name', 'Span', 'Innings', 'Runs', 'Batting_Average']
 thewriter.writerow(header)

 for list in lists:
  name = list.find('a', class_="data-link").text
  span = list.find('td', class_="left").text
  innings = list.find('td', class_="padAst").text
  runs = list.find('b').text
  battingg_avg = list.find('td', "nowrap")
  
  info = [name, span, innings, runs, battingg_avg] 
  thewriter.writerow(info)
```
